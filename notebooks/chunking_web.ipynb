```python
import json
val_web = json.load(open('./val_selected_chunks.json'))
test_web = json.load(open('./test_selected_chunks.json'))
corpus = []

for item in val_web:
    title = item['question'].strip().lower()
    all_chunks = []
    for page in item['chunks_page']:
        all_chunks.extend(page)
    assert all(isinstance(d, str) for d in all_chunks)
    for chunk_id, chunk in enumerate(all_chunks):
        chunk_name = f'{title}_chunk{chunk_id}'
        corpus.append({
            'title': chunk_name,
            'content': f"Tiêu đề: {title}\n{chunk}",
        })


for item in test_web:
    title = item['question'].strip().lower()
    all_chunks = []
    for page in item['chunks_page']:
        all_chunks.extend(page)
    assert all(isinstance(d, str) for d in all_chunks)
    for chunk_id, chunk in enumerate(all_chunks):
        chunk_name = f'{title}_chunk{chunk_id}'
        corpus.append({
            'title': chunk_name,
            'content': f"Tiêu đề: {title}\n{chunk}",
        })

```

```python
import json
json.dump(corpus, open('./corpus/chunks_web.json', 'w', encoding='utf-8'), ensure_ascii=False, indent=2)
```
